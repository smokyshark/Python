#### Setup

```python
# import matplotlib and pandas
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline

# load datasets
bar_data = pd.read_csv('plant_data_simplified.csv')
scatter_data = pd.read_csv('spotify_data_by_genres.csv')
pie_data = pd.read_csv('EV_market_share.csv')
hist_data = pd.read_csv('lobster_dist.csv')
```

```python
# display charts correctly
%matplotlib inline
plt.rcParams['figure.figsize'] = (5, 3)
plt.rcParams['figure.dpi'] = 75
```

1. Bar chart & error bars

```python
plt.bar(x = bar_data.PH, height = bar_data.average_leaf_width, width = 0.8, align = 'center')
plt.title('Effect of pH on Leaf Width')
plt.xlabel('pH')
plt.ylabel('Leaf Width (cm)')
plt.errorbar(bar_data.PH, bar_data.average_leaf_width, yerr = bar_data.error, fmt = 'o', color = 'orangered')
plt.show()
```

2. Scatterplot & subplots

```python
plt.suptitle('Relationship between Popularity and ______')
plt.subplot(2,2,1)
plt.scatter(scatter_data.popularity, scatter_data.acousticness, alpha=0.05)
plt.title('acousticness')
plt.subplot(2,2,2)
plt.scatter(scatter_data.popularity, scatter_data.danceability, alpha=0.05)
plt.title('danceability')
plt.subplot(2,2,3)
plt.scatter(scatter_data.popularity, scatter_data.loudness, alpha=0.05)
plt.title('loudness')
plt.subplot(2,2,4)
plt.scatter(scatter_data.popularity, scatter_data.tempo, alpha=0.05)
plt.title('tempo')
plt.subplots_adjust(hspace=0.3, wspace=0.3)
plt.rcParams['figure.figsize'] = (5, 5) #this adjusts the figure size so the graph fits better
plt.show()
```

3. Pie chart

```python
plt.pie(pie_data.number_sold, labels=pie_data.model, startangle=60)
plt.title('Top 10 EVs sold in the US by count, 2021')
plt.show()
```

4. Histogram with AB lines & annotations

```python
plt.hist(hist_data.carapace_length, bins=30, range=(75,155), color='gold')
plt.title('Lobsters tagged by size')
plt.ylabel('Number tagged')
plt.xlabel('Carapace length (mm)')
plt.axline((83, 0), (83, 1), linewidth=2, dashes=(1,2), color='mediumblue')
plt.axline((127, 0), (127, 1), linewidth=2, dashes=(1,2), color='mediumblue')
plt.annotate('Minimum legal catch size', (84, 23), color='mediumblue')
plt.annotate('Maximum legal\ncatchsize', (128, 7), color='mediumblue')
plt.show()
```
