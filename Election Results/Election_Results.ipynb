```python
import numpy as np
from matplotlib import pyplot as plt


survey_responses = ['Ceballos', 'Kerrigan', 'Ceballos', 'Ceballos', 'Ceballos','Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Ceballos', 
'Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Ceballos', 'Ceballos', 'Ceballos', 'Ceballos',
'Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Ceballos', 'Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Ceballos',
'Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Ceballos', 'Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Kerrigan', 'Kerrigan', 'Ceballos', 'Ceballos', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Kerrigan', 'Ceballos']

total_ceballos = sum([1 for response in survey_responses if response == 'Ceballos'])
total_kerrigan = sum([1 for i in survey_responses if i == 'Kerrigan'])

print('Cynthia Ceballos votes: ', total_ceballos)
print('Kerrigan votes: ', total_kerrigan)

survey_length = len(survey_responses)
percentage_ceballos = total_ceballos / survey_length * 100
print('Percentage Cynthia Ceballos votes: ', round(percentage_ceballos, 2), '%')

percentage_kerrigan = total_kerrigan / survey_length * 100
print('Percentage Kerrigan votes: ', round(percentage_kerrigan, 2), '%')


#In real world Cybthia Ceballos got 54% of votes
possible_surveys = np.random.binomial(survey_length, 0.54, size = 10000) / survey_length

ceballos_loss_surveys = np.mean(possible_surveys < 0.5)
print('Ceballos loss surveys: ', ceballos_loss_surveys)

```

```python
#
large_survey = np.random.binomial(7000, 0.54, size = 10000) / 7000
```

```python
plt.hist(possible_surveys, range = (0,1), bins = 20, label = 'Possible surveys')
plt.hist(large_survey, alpha = 0.5, range = (0, 1), bins = 20, label = 'Large survey')
plt.title('Election Results')
plt.xlabel('Percentage of votes')
plt.ylabel('Population')
plt.legend(loc=2)
plt.show() 
```

```python

ceballos_loss_new = np.mean(large_survey < 0.5)
print('There is ', ceballos_loss_new, '% chance to get result below 50% when using sample size og 7000.')
```
